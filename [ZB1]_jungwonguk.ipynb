### 아쉬운점

- 각 매장별 좌표값을 제대로 가져왔다면 지도로 데이터 시각화를 통해 비교가 가능했을텐데 아쉬움이 있다.
- 또한 후에 html 을 뜯어보니 <li class="quickResultLstCon" style="background:#fff" data-lat="37.501087" data-long="127.043069" 이런식으로 좌표가 있었는데 이걸 떼서 칼럼에 넣어서 분석할걸 잘못했다는 생각이 든다.

#### 가정
- 마지막으로 제대로 좌표값이 들어가고 지도에 데이터를 시각해서 표현했다고 가정해보자
- 각 구별로 스타벅스와 이디아가 얼마나 분포되었는지를 알 수 있을것이다.
- 하지만 과연 이렇게 지도에 붙어있다고 해서 스타벅스를 이디아가 따라갔다고 할 수 있는가?
- 이점은 어떤 매장이 먼저 지어졌는지에 대한 데이터가 추가로 있어야만 연관성을 찾을 수 있다고 본다.
- 추가적으로 각 구의 거리별 유동인구수에 대한 데이터가 나온다면 단순히 각 커피 브랜드 매장은 유동 인구의 데이터를 따라갔다고도 분석할 수 있지 않을까 생각해봤다.
- 위의 가정은 추후에 더 진행해봐야겠다.
